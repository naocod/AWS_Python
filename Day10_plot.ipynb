```python
print(np.pi)
print(np.exp(5))
print(np.sqrt(25))

#변수
mu = 0      # 평균
sig = 1     # 표준편차, 분산의 제곱근
x = np.linspace(-5, 5, 100)
y = 1 / (sig * np.sqrt(2 * np.pi))
# 표준 정규 분포
y = y + np.exp(-(x - mu) ** 2) / (2 * sig * sig)

_, axe = plt.subplots()
axe.plot(x, y)
```

```python
import math

mu = 0
sig = 1
x = np.linspace(-5, 5, 100)
y = []
for v in x:
    y.append(0.5 * (1 + math.erf((v - mu) / (sig * np.sqrt(2)))))
# y = 0.5 * (1 + math.erf((x - mu) / (sig * np.sqrt(2))))
# print(y)

_, axe = plt.subplots()
axe.plot(x, y)
```
