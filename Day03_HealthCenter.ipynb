```python
# 빈도수 dictionary 반환
# 1. raw_list를 set으로 변환
# 2. set을 순환하면서, 각각 몇개씩 raw_list에 있는지 카운트
# 3. 1,2에서 얻은 key, value 값으로 dictionary 만들어서 return

# lambda x: x[1]
def sort_key(x):
    return x[1]

raw_count = []
def unique_count(raw_list, sort = False): # sort = False 오름차순, True 내림차순
    raw_set = set(raw_list)
    k_list = []
    v_list = []
    for i in raw_set:
        k_list.append(i)
        v_list.append(raw_list.count(i))

    ret_dict = dict(zip(k_list, v_list))
    if sort == True:
        # ret_dict = dict(sorted(ret_dict.items(), key = lambda x: x[1]))
        ret_dict = dict(sorted(ret_dict.items(), key = sort_key)) # 람다 사용하지 않고 변수로 sort
    return ret_dict

unique_count(name_list)
```
