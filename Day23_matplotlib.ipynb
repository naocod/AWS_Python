```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

# print(xs)
# print(ys)

figure, axe = plt.subplots()
figure.set_facecolor('skyblue')
figure.suptitle('Figurel')
figure.supxlabel('Figurel_x')
figure.supylabel('Figurel_y')
figure.text(0.5, 0.7, 'mytext')
figure.set_size_inches(15, 6.8)

axe.plot(xs, ys)
axe.set_title('my axe title')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')

axe.set_xlim(0, 30)
axe.set_ylim(0, 1)

axe.set_xticks(np.linspace(0, 30, 5))
axe.set_xticklabels(list('abcde'))

print(type(figure))
print(type(axe))
```
