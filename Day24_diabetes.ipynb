```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = "https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt"
pd_data = pd.read_csv(url, sep='\t')
print(pd_data)
```

```python
_, axe = plt.subplots()
# axe.plot(pd_data['BMI'])
axe.hist(pd_data['AGE'], bins=20, linewidth=2.0, edgecolor="skyblue")
```

```python
_, ax = plt.subplots()
ax.boxplot(pd_data['AGE'])
pd_data['AGE'].describe()
```
