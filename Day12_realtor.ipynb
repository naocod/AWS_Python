### 전국공인중개사 사무소표준데이터
    - 공인중개사 / 개업공인중개사 비율
    - 공제미가입 중개사 count, 전체 대비 비율
    - ㅇㅇ공인중개... 중 흔한 상호명
    - ㅇㅇ공인중개...가 아닌 것은 따로 저장

```python
import numpy as np

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list = []
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list)) > 1:
        for i in set(len_list):
            print(i, len_list.count(i))    

raw_data = []
with open('전국공인중개사사무소표준데이터.csv') as f:
    for line in f:
        raw_data.append(my_split(line[:-1]))
print(raw_data[:3])
split_len(raw_data)
```
