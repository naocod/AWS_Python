```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd_tips = sns.load_dataset('tips')
print(pd_tips)
```

```python
sns.catplot(data = pd_tips, x = 'day', y = 'total_bill', jitter=0.2, hue='smoker', dodge=True) # jitter를 사용하면 relplot과 같은 결과가 나옴
# sns.relplot(data = pd_tips, x = 'day', y = 'total_bill')
```

```python
sns.catplot(data = pd_tips, x = 'day', y = 'total_bill', hue='smoker', dodge=True, kind='swarm')
```

```python
pd_tips.plot(x='day', y='total_bill', kind='line')
```

```python
pd_tips.plot.line(x='day', y='total_bill')
```
