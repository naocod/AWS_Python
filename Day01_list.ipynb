```python
list_a = [1, 2, 3]
list_b = list('abc')
d1 = {'a' : 1, 'b' : 2, 'c' : 3}
print(list_a)
print(list_b)
print(d1)

#              key  ,  value
d2 = dict(zip(list_b, list_a)) # 두 개의 리스트를 key와 value로 받아서 dict 타입으로 변환
print(d2)
print(d2.items())

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

print(d3['a']) # key값으로 value조회 없는 키값을 조회하면 에러뜸
print(d3.get('c')) # key값으로 조회하는것과 같지만 없는 키값으로 조회하면 에러가 나지 않고 return값이 없어 none으로 출력됨
d3['y'] = 15
print(d3)
```

```python
for i in d3:
    print(i) # 키값만 나옴
    
for i in d3:
    print(i, d3[i]) # key value 조회
    
print(d3.keys()) # key
print(d3.values()) # value

for i in d3.values():
    print(i)
    
print(d3)
d3['y'] = 20
print(d3)
```
